#1.line chart


```python
import matplotlib.pyplot as plt
nums=[0,1,2,3,4,5]
squares=[i**2 for i in nums]
mult_of_5=[i*5 for i in nums]
plt.plot(nums,mult_of_5,c='r',marker="d",mec='b',mfc='g',ms='10')
plt.plot(nums,squares,marker="*",mec='r',mfc='c',ms='15')
plt.title("square numbers",fontsize=24)
plt.xlabel("value",fontsize=14)
plt.ylabel("square of value",fontsize=14)
plt.tick_params(axis='both',labelsize=14)
plt.grid()
plt.show()
```

subplots

```python
import numpy as np
x=np.arange(0,9,2)
a=[10,20,40,50,30]
b=[50,10,30,40,20]
c=[10,20,30,40,50]
d=[40,30,50,10,20]
plt.suptitle("subplot")
plt.subplot(2,2,1)
plt.plot(x,a,c='r',marker="d",mec='b',mfc='g',ms='10')
plt.title("plot-a")

plt.subplot(2,2,2)
plt.plot(x,b,c='b',marker="*",mec='r',mfc='c',ms='15')
plt.title("plot-b")

plt.subplot(2,2,3)
plt.plot(x,c,c="g",marker=".")
plt.title("plot-c")

plt.subplot(2,2,4)
plt.plot(x,d,c="k",marker=",")
plt.title("plot-d")

plt.show()
```

scatter() is used to create a scatter plot & displaying individual data points on a 2d graph

```python
import random
a=np.arange(1,11)
b=np.arange(10,101,10)
print(a,b)
plt.scatter(a,b)
random.shuffle(b)
print(a,b)
plt.scatter(a,b,marker="*",color='r',alpha=0.6)
plt.show()
```

another way of using plt.scatter()
-> plt.colorbar() adds a color bar to the plot,showing the mapping from data values to colors.

```python
x=[10,20,30,40,50]
y=[200,300,100,400,500]
colors=['r','g','b','c','m']
sizes=[150,200,300,250,350]
plt.scatter(x,y,c=colors,s=sizes)
plt.colorbar()
plt.show()
# try with different types of colormaps & check
```

```python
import pandas as pd
data=pd.read_excel("India_2024_Lok_Sabha_Results.xlsx")
print(pd.DataFrame(data))
```


bold text extract the parties and seats lists from dataframe

```python
# Extract list of parties
parties = data['Party Name'].tolist()

# Extract list of seats
seats = data['Seats Won'].tolist()
```

```python
# Plot pie chart
plt.figure(figsize=(15, 7))
plt.pie(seats, labels=parties, autopct='%1.1f%%', startangle=90)
plt.title("2024 Lok Sabha Seats Distribution")
plt.axis('equal')  # Makes the pie a circle
plt.legend(facecolor="c")
plt.show()
```
