#1.line chart


```python
import matplotlib.pyplot as plt
nums=[0,1,2,3,4,5]
squares=[i**2 for i in nums]
mult_of_5=[i*5 for i in nums]
plt.plot(nums,mult_of_5,c='r',marker="d",mec='b',mfc='g',ms='10')
plt.plot(nums,squares,marker="*",mec='r',mfc='c',ms='15')
plt.title("square numbers",fontsize=24)
plt.xlabel("value",fontsize=14)
plt.ylabel("square of value",fontsize=14)
plt.tick_params(axis='both',labelsize=14)
plt.grid()
plt.show()
```

subplots

```python
import numpy as np
x=np.arange(0,9,2)
a=[10,20,40,50,30]
b=[50,10,30,40,20]
c=[10,20,30,40,50]
d=[40,30,50,10,20]
plt.suptitle("subplot")
plt.subplot(2,2,1)
plt.plot(x,a,c='r',marker="d",mec='b',mfc='g',ms='10')
plt.title("plot-a")

plt.subplot(2,2,2)
plt.plot(x,b,c='b',marker="*",mec='r',mfc='c',ms='15')
plt.title("plot-b")

plt.subplot(2,2,3)
plt.plot(x,c,c="g",marker=".")
plt.title("plot-c")

plt.subplot(2,2,4)
plt.plot(x,d,c="k",marker=",")
plt.title("plot-d")

plt.show()
```

```python
#Grouped bar graphs
import numpy as np
import matplotlib.pyplot as plt
years=["2021","2022","2023","2024","2025"]
bca=[80,75,85,90,95]
bba=[10,12,14,18,8]
bcom=[35,30,40,37,44]
bsc=[38,30,60,52,55]
w=0.2
plt.figure(figsize=(10,6))
#location of the bar
bca_bar=np.arange(len(years))
bba_bar=bca_bar+w
bcom_bar=bba_bar+w
bsc_bar=bcom_bar+w
#plotting each groups data
plt.bar(bca_bar,bca,width=w,label="BCA",color="yellow")
plt.bar(bba_bar,bba,width=w,label="BBA",color="g")
plt.bar(bcom_bar,bcom,width=w,label="BCOM",color="b")
plt.bar(bsc_bar,bsc,width=w,label="BSC",color="r")
#adding labels,title & legend
plt.xticks(bca_bar+w+w/2,years)
plt.ylim(0,100)
plt.xlabel("years")
plt.ylabel("placements")
plt.title("Placement Analysis")
plt.legend()
plt.show()
```

#3.pie charts

analysing the students performance in a school

```python
import matplotlib.pyplot as plt
performance=["exelent","good","average","poor"]
color=["g","m","y","r"]
merits=[9,18,27,46]
plt.figure(figsize=(10,8))
plt.pie(merits,labels=performance,explode=[0.4,0,0,0],shadow=True,colors=color,autopct="%1.2f%%",frame=True)
plt.title("Students Performance")
plt.show()
```

#4.histogram charts

```python
import matplotlib.pyplot as plt
import numpy as np
marks=np.random.randint(0,100,size=12)
print(marks)
grade_interval=[0,35,50,75,100]
plt.title("student grades")
plt.hist(marks,grade_interval,histtype="barstacked",rwidth=0.8,color="greenyellow")
# histtype = 'bar','step','stepfilled','barstacked'
plt.xticks(grade_interval)
plt.xlabel("marks")
plt.ylabel("frequency")
plt.show()
```

#5.scatter plot

->analysing economic indicators : relationship b/w per capita
->NSDP and projected GSDP with population for indian states

```python
# Analyzing Economic indicators: relationship law per capita
# NSDP and projected GSDP with population for nation states
import matplotlib.pyplot as plt
states = ["maharasota", "Tamil road", "Gujarat", "Kamadaka", "utlas Pradesh"]

gscdp = [38.19, 28.7, 25.62, 25, 24.39] # RSI in Lakh score
per_capita_nscdp = [2.21, 2.13, 2.41, 3.0, 0.83] # RSI in Lakh
population = [11, 7, 6, 6.1, 19.9] # Population in scores
colors=['r','g','b','c','m']
# Create a scatter plot
plt.figure(figsize = (12,8))
plt.scatter(per_capita_nscdp, gscdp, s = [pop*50 for pop in population], color = colors, edgecolor = 'r', alpha = 0.7)

plt.title("per capita NSDP vs. GSDP for Top 5 states of India in 2023-24", fontsize = 18)
plt.xlabel("per capita NSDP (zlab) in (2022-23 FY)", fontsize = 14)
plt.ylabel("GSDP (zlab score) (FY 2023-24)", fontsize = 14)

# Add annotations for each state
for i, state in enumerate(states):
    plt.annotate(state, (per_capita_nscdp[i], gscdp[i]), fontsize = 12, ha = 'left')

plt.grid(True)
#plt.legend(title="size of scatters based on population", loc='center', facecolor = 'yellow')
plt.legend()
plt.show()
```
